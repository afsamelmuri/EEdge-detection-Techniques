**1. Sobel Edge Detection**

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread(r'C:\Users\Lenovo\Downloads\daimond pp\2nd sem video lab\Lena.jfif', cv2.IMREAD_GRAYSCALE)


# Sobel operator in the x direction
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)

# Sobel operator in the y direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Converting back to uint8
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)

# Combine Sobel X and Y
sobel_combined = cv2.bitwise_or(sobel_x, sobel_y)

# Display the results
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(sobel_x, cmap='gray'), plt.title('Sobel X')
plt.subplot(1,3,2), plt.imshow(sobel_y, cmap='gray'), plt.title('Sobel Y')
plt.subplot(1,3,3), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel Combined')
plt.show()
```

**2. Prewitt Edge Detection**

```python

# Prewitt edge detection kernels
kernelx = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
kernely = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

prewitt_x = cv2.filter2D(image, -1, kernelx)
prewitt_y = cv2.filter2D(image, -1, kernely)

# Combine the two
prewitt_combined = cv2.bitwise_or(prewitt_x, prewitt_y)

# Display the results
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(prewitt_x, cmap='gray'), plt.title('Prewitt X')
plt.subplot(1,3,2), plt.imshow(prewitt_y, cmap='gray'), plt.title('Prewitt Y')
plt.subplot(1,3,3), plt.imshow(prewitt_combined, cmap='gray'), plt.title('Prewitt Combined')
plt.show()
```

**3. Canny Edge Detection**

```python

# Apply Gaussian blur to reduce noise
blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)

# Apply Canny edge detection
edges = cv2.Canny(blurred_image, threshold1=50, threshold2=150)

# Display the result
plt.figure(figsize=(6,6))
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.show()
```

**4. Laplacian Edge Detection**

```python

# Apply Laplacian operator
laplacian = cv2.Laplacian(image, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)

# Display the result
plt.figure(figsize=(6,6))
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Edge Detection')
plt.show()
```

**5. Roberts Edge Detection**

```python

# Roberts operator kernels
kernelx = np.array([[1, 0], [0, -1]])
kernely = np.array([[0, 1], [-1, 0]])

roberts_x = cv2.filter2D(image, -1, kernelx)
roberts_y = cv2.filter2D(image, -1, kernely)

# Combine the two
roberts_combined = cv2.bitwise_or(roberts_x, roberts_y)

# Display the results
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(roberts_x, cmap='gray'), plt.title('Roberts X')
plt.subplot(1,3,2), plt.imshow(roberts_y, cmap='gray'), plt.title('Roberts Y')
plt.subplot(1,3,3), plt.imshow(roberts_combined, cmap='gray'), plt.title('Roberts Combined')
plt.show()
```

**6. Scharr Edge Detection**

```python

# Scharr operator in the x and y directions
scharr_x = cv2.Scharr(image, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(image, cv2.CV_64F, 0, 1)

# Converting back to uint8
scharr_x = cv2.convertScaleAbs(scharr_x)
scharr_y = cv2.convertScaleAbs(scharr_y)

# Combine the two
scharr_combined = cv2.bitwise_or(scharr_x, scharr_y)

# Display the results
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(scharr_x, cmap='gray'), plt.title('Scharr X')
plt.subplot(1,3,2), plt.imshow(scharr_y, cmap='gray'), plt.title('Scharr Y')
plt.subplot(1,3,3), plt.imshow(scharr_combined, cmap='gray'), plt.title('Scharr Combined')
plt.show()
```

**7. Zero-Crossing Based Edge Detection (LoG)**

```python

# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)

# Apply Laplacian operator
log = cv2.Laplacian(blurred_image, cv2.CV_64F)
log = cv2.convertScaleAbs(log)

# Display the result
plt.figure(figsize=(6,6))
plt.imshow(log, cmap='gray')
plt.title('Laplacian of Gaussian (LoG)')
plt.show()
```

**8. Difference of Gaussians (DoG)**

```python

# Apply Gaussian blurs with different sigma values
blur1 = cv2.GaussianBlur(image, (5, 5), 1)
blur2 = cv2.GaussianBlur(image, (5, 5), 2)

# Subtract the two blurred images
dog = blur1 - blur2

# Display the result
plt.figure(figsize=(6,6))
plt.imshow(dog, cmap='gray')
plt.title('Difference of Gaussians (DoG)')
plt.show()
```
